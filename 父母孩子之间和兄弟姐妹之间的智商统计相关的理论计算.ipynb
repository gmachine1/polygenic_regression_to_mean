```
var('rho,t,x,y,y1,y2')
rho = 0.67
sigma = sqrt(1.0-rho^2)

y_pdf = (1/(sigma*sqrt(2*pi)))*e^(-0.5*(y-rho*x)^2/sigma^2)
x_pdf = (1/sqrt(2*pi))*e^(-x^2/2)

parent_child_integrand = x*y*x_pdf*y_pdf
sibling_integrand = y1*y2*x_pdf*y_pdf.subs(y==y1)*y_pdf.subs(y==y2)
sibling_integrand
```

```
# integral(integral(parent_child_integrand, (x,-5,5)), (y,-5,5)) throws an error
inc = 0.05
X = -5
parent_child_corr = 0.0
ct = 0
while X <= 5:
    Y = max(X-4, -5)
    while Y <= min(5, X+4):
        parent_child_corr += parent_child_integrand.subs(x==X,y==Y)*inc^2
        Y += inc
        ct += 1
        if ct % 1000 == 0:
            print(ct)
            print (str(X) + " " + str(Y))
    X += inc
numerical_approx(parent_child_corr)
```
